```python
# EDA summary & Reporting

# cancellation rate by hotel type

cancellation_summary = (
    df.groupby("hotel")["is_canceled"]
      .agg(
          Total_Bookings="count",
          Cancellation_Rate="mean"
      )
)

cancellation_summary["Cancellation_Rate"] = (
    cancellation_summary["Cancellation_Rate"] * 100
).round(2)

cancellation_summary
```

```python
# ADR summary by hotel type

adr_summary = (
    df.groupby("hotel")["adr"]
      .agg(
          Average_ADR="mean",
          Minimum_ADR="min",
          Maximum_ADR="max"
      )
      .round(2)
)

adr_summary
```

```python
# customer type summary

customer_summary = (
    df.groupby("customer_type")
      .agg(
          Total_Bookings=("customer_type","count"),
          Average_ADR=("adr","mean"),
          Cancellation_Rate=("is_canceled","mean")
      )
      .round(2)
)

customer_summary["Cancellation_Rate"] *= 100

customer_summary
```

```python
# market segment summary

market_summary = (
    df.groupby("market_segment")
      .agg(
          Bookings=("market_segment","count"),
          Average_ADR=("adr","mean"),
          Cancellation_Rate=("is_canceled","mean")
      )
      .round(2)
)

market_summary["Cancellation_Rate"] *= 100

market_summary.sort_values(
    "Bookings",
    ascending=False
)
```

```python
# top 10 countries summary

country_summary = (
    df.groupby("country")
      .agg(
          Bookings=("country","count"),
          Average_ADR=("adr","mean")
      )
      .round(2)
      .sort_values(
          "Bookings",
          ascending=False
      )
      .head(10)
)

country_summary
```

```python
cancellation_summary.to_csv(
    "../reports/cancellation_summary.csv"
)

adr_summary.to_csv(
    "../reports/adr_summary.csv"
)

customer_summary.to_csv(
    "../reports/customer_summary.csv"
)

market_summary.to_csv(
    "../reports/market_summary.csv"
)

country_summary.to_csv(
    "../reports/country_summary.csv"
)
```
