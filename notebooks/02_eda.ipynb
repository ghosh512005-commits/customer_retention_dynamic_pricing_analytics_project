```python
import pandas as pd

df = pd.read_csv("../data/cleaned/hotel_bookings_cleaned.csv")

df.head()
```

```python
# import libraries

import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("ggplot")
sns.set_theme(style="whitegrid")

# df.shape
# df.info()
# df.describe()
df.isnull().sum()
```

```python
# hotel type distribution

plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="hotel"
)
plt.savefig(
    "../screenshots/week2/hotel_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.title("Hotel Type Distribution")
plt.show()
```

```python
# cancellation distribution

plt.figure(figsize=(6,4))

sns.countplot(
    data=df,
    x="is_canceled"
)

plt.title("Booking Cancellation Distribution")

plt.savefig(
    "../screenshots/week2/cancellation_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
# booking distribution by month

plt.figure(figsize=(12,5))

sns.countplot(
    data=df,
    x="arrival_date_month",
    order=[
        "January","February","March","April","May","June",
        "July","August","September","October","November","December"
    ]
)

plt.xticks(rotation=45)

plt.title("Bookings by Month")

plt.savefig(
    "../screenshots/week2/monthly_bookings.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```

```python
# adr distribution

plt.figure(figsize=(8,5))

sns.histplot(
    df["adr"],
    bins=40,
    kde=True
)

plt.title("Distribution of Average Daily Rate")

plt.savefig(
    "../screenshots/week2/adr_distribution.png",
    dpi=300,
    bbox_inches="tight"
)

plt.show()
```
